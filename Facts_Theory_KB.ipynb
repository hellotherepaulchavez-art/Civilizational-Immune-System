```python
import requests
from bs4 import BeautifulSoup
import re

class WebNavigator:
    def __init__(self, timeout=10):
        self.timeout = timeout
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }

    def scrape_url(self, url):
        """Fetches a URL and returns a clean text representation."""
        try:
            response = requests.get(url, headers=self.headers, timeout=self.timeout)
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')

            # Remove script and style elements
            for script in soup(["script", "style"]):
                script.extract()

            text = soup.get_text(separator=' ')
            # Clean whitespace
            lines = (line.strip() for line in text.splitlines())
            chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
            text = '\n'.join(chunk for chunk in chunks if chunk)

            return {"url": url, "content": text[:5000], "title": soup.title.string if soup.title else "No Title"}
        except Exception as e:
            return {"url": url, "error": str(e)}

    def extract_links(self, text):
        """Finds all HTTP/S links in a block of text."""
        return re.findall(r'(https?://\S+)', text)

nav = WebNavigator()
print("WebNavigator tool initialized.")
```

```python
import os, json, sqlite3, hashlib, yaml
from pathlib import Path
from datetime import datetime

def now():
    return datetime.utcnow().isoformat() + "Z"

def read_text(path):
    return Path(path).read_text(encoding="utf-8")

def load_json(path):
    return json.loads(read_text(path))

def load_yaml(path):
    return yaml.safe_load(read_text(path))

def upsert_item(cur, item):
    cur.execute("""
    INSERT INTO items (id, item_type, title, body, topic, status, confidence, citations, tags, source_path, created_at, updated_at)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ON CONFLICT(id) DO UPDATE SET
        item_type=excluded.item_type,
        title=excluded.title,
        body=excluded.body,
        topic=excluded.topic,
        status=excluded.status,
        confidence=excluded.confidence,
        citations=excluded.citations,
        tags=excluded.tags,
        source_path=excluded.source_path,
        updated_at=excluded.updated_at
    """, (
        item["id"],
        item["item_type"],
        item["title"],
        item["body"],
        item.get("topic"),
        item.get("status", "draft"),
        float(item.get("confidence", 0.5)),
        json.dumps(item.get("citations", [])),
        json.dumps(item.get("tags", [])),
        item.get("source_path"),
        item.get("created_at", now()),
        now(),
    ))

def upsert_source(cur, src):
    cur.execute("""
    INSERT INTO sources (id, name, url, source_type, reliability, notes, created_at, updated_at)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    ON CONFLICT(id) DO UPDATE SET
        name=excluded.name,
        url=excluded.url,
        source_type=excluded.source_type,
        reliability=excluded.reliability,
        notes=excluded.notes,
        updated_at=excluded.updated_at
    """, (
        src["id"],
        src["name"],
        src.get("url"),
        src.get("source_type"),
        src.get("reliability"),
        src.get("notes"),
        src.get("created_at", now()),
        now(),
    ))

def log(cur, action, item_id=None, path=None, detail=None):
    cur.execute("""
    INSERT INTO audit_log (action, item_id, path, detail, created_at)
    VALUES (?, ?, ?, ?, ?)
    """, (action, item_id, path, detail, now()))

def load_item_file(path):
    path = Path(path)
    ext = path.suffix.lower()
    data = None

    if ext == ".json":
        data = load_json(path)
    elif ext in [".yml", ".yaml"]:
        data = load_yaml(path)
    elif ext == ".md":
        text = read_text(path)
        if text.startswith("---"):
            parts = text.split("---", 2)
            if len(parts) >= 3:
                fm = parts[1]
                body = parts[2]
                meta = yaml.safe_load(fm) or {}
                data = {
                    "id": meta["id"],
                    "item_type": meta["item_type"],
                    "title": meta["title"],
                    "body": body.strip(),
                    "topic": meta.get("topic"),
                    "status": meta.get("status", "draft"),
                    "confidence": meta.get("confidence", 0.5),
                    "citations": meta.get("citations", []),
                    "tags": meta.get("tags", []),
                    "source_path": str(path),
                }
        else:
            raise ValueError(f"Markdown missing frontmatter: {path}")
    else:
        return None

    return data

def scan_and_load(root):
    conn = sqlite3.connect(DB_PATH)
    cur = conn.cursor()

    for p in Path(root).rglob("*"):
        if not p.is_file():
            continue
        if p.suffix.lower() not in [".json", ".yaml", ".yml", ".md"]:
            continue

        try:
            data = load_item_file(p)
            if not data:
                continue
            data["source_path"] = str(p)

            if "item_type" in data and "title" in data and "body" in data:
                upsert_item(cur, data)
                log(cur, "load_item", item_id=data["id"], path=str(p))
            elif "name" in data and "source_type" in data:
                upsert_source(cur, data)
                log(cur, "load_source", path=str(p), detail=data["id"])
            else:
                print("Unrecognized record:", p)
        except Exception as e:
            print(f"Skip {p}: {e}")

    conn.commit()
    conn.close()
    print("Load complete.")
```

```python
import sqlite3, json
from pathlib import Path
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def load_items():
    conn = sqlite3.connect(DB_PATH)
    cur = conn.cursor()
    cur.execute("""
        SELECT id, item_type, title, body, topic, status, confidence, citations, tags, source_path
        FROM items
    """)
    rows = cur.fetchall()
    conn.close()

    items = []
    for r in rows:
        items.append({
            "id": r[0],
            "item_type": r[1],
            "title": r[2],
            "body": r[3],
            "topic": r[4],
            "status": r[5],
            "confidence": r[6],
            "citations": json.loads(r[7] or "[]"),
            "tags": json.loads(r[8] or "[]"),
            "source_path": r[9],
        })
    return items

def item_text(item):
    return "\n".join([
        item.get("title",""),
        item.get("body",""),
        f"topic: {item.get('topic','')}",
        f"tags: {', '.join(item.get('tags', []))}",
        f"citations: {', '.join(item.get('citations', []))}",
        f"type: {item.get('item_type','')}",
    ])

class KBRetriever:
    def __init__(self, items):
        self.items = items
        self.vectorizer = TfidfVectorizer(stop_words="english", max_features=20000)
        self.matrix = self.vectorizer.fit_transform([item_text(i) for i in items]) if items else None

    def search(self, query, top_k=5):
        if not self.items or self.matrix is None:
            return []
        q_vec = self.vectorizer.transform([query])
        sims = cosine_similarity(q_vec, self.matrix).flatten()
        ranked = sorted(enumerate(sims), key=lambda x: x[1], reverse=True)
        results = []
        for idx, score in ranked[:top_k]:
            item = self.items[idx].copy()
            item["score"] = float(score)
            results.append(item)
        return results

def build_context(results):
    blocks = []
    for r in results:
        blocks.append(f"""[ID: {r['id']} | TYPE: {r['item_type']} | SCORE: {r['score']:.3f}]
Title: {r['title']}
Body: {r['body']}
Citations: {", ".join(r.get("citations", [])) or "none"}
Source: {r.get("source_path")}
""")
    return "\n---\n".join(blocks)

items = load_items()
retriever = KBRetriever(items)

print("Items loaded:", len(items))
```

```python
retriever_code = """
import os
import json
import sqlite3
from pathlib import Path
from typing import List, Dict, Any
import yaml
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

KB_ROOT = "/content/drive/MyDrive/facts-theory-kb"
DB_PATH = Path(KB_ROOT, "db", "kb.sqlite3")
TEXT_EXTS = {".md", ".json", ".yaml", ".yml"}

def read_record_file(path: Path) -> Dict[str, Any]:
    if path.suffix.lower() == ".json":
        return json.loads(path.read_text(encoding="utf-8"))
    if path.suffix.lower() in {".yaml", ".yml"}:
        return yaml.safe_load(path.read_text(encoding="utf-8"))
    if path.suffix.lower() == ".md":
        text = path.read_text(encoding="utf-8")
        if text.startswith("---"):
            parts = text.split("---", 2)
            if len(parts) >= 3:
                fm = parts[1]
                body = parts[2]
                meta = yaml.safe_load(fm) or {}
                return {
                    "id": meta.get("id"),
                    "item_type": meta.get("item_type"),
                    "title": meta.get("title"),
                    "body": body.strip(),
                    "topic": meta.get("topic"),
                    "status": meta.get("status", "draft"),
                    "confidence": meta.get("confidence", 0.5),
                    "citations": meta.get("citations", []),
                    "tags": meta.get("tags", []),
                    "source_path": str(path),
                }
    return {}

def load_items_from_files(root: str) -> List[Dict[str, Any]]:
    items = []
    for p in Path(root).rglob("*"):
        if p.is_file() and p.suffix.lower() in TEXT_EXTS:
            try:
                data = read_record_file(p)
                if data and data.get("id") and data.get("title"):
                    items.append(data)
            except Exception as e:
                print(f"Skip {p}: {e}")
    return items

class KBRetriever:
    def __init__(self, items: List[Dict[str, Any]]):
        self.items = items
        self.vectorizer = TfidfVectorizer(stop_words="english", max_features=20000)
        self.matrix = self.vectorizer.fit_transform([self.item_text(i) for i in items]) if items else None

    def item_text(self, item: Dict[str, Any]) -> str:
        return f"{item.get('title', '')} {item.get('body', '')} {item.get('topic', '')}"

    def search(self, query: str, top_k: int = 5):
        if not self.items or self.matrix is None:
            return []
        q_vec = self.vectorizer.transform([query])
        sims = cosine_similarity(q_vec, self.matrix).flatten()
        ranked = sorted(enumerate(sims), key=lambda x: x[1], reverse=True)
        return [(self.items[idx], score) for idx, score in ranked[:top_k] if score > 0]
"""

with open(os.path.join(KB_ROOT, "scripts/retriever.py"), "w") as f:
    f.write(retriever_code)
print("Retriever script saved to Drive.")
```

## Define Autonomous Loop Controller

### Subtask:
Create the `AutonomousLoopController` class to orchestrate the research workflow and manage state across iterations.


**Reasoning**:
I will define the `AutonomousLoopController` class to manage the research loop, integrating the navigation, retrieval, and storage components as instructed.



```python
import time

class AutonomousLoopController:
    def __init__(self, nav, retriever, kb_root):
        self.nav = nav
        self.retriever = retriever
        self.kb_root = kb_root
        self.research_goal = ""
        self.visited_urls = set()
        self.link_queue = []
        self.iteration_history = []

    def set_goal(self, goal):
        self.research_goal = goal
        print(f"Research Goal Set: {self.research_goal}")

    def run_iteration(self, target_url=None):
        """Executes a single cycle: Goal -> Search -> Scrape -> Analyze -> Save"""
        try:
            # 1. Selection: If no URL provided, search KB or use queue
            if not target_url:
                if self.link_queue:
                    target_url = self.link_queue.pop(0)
                else:
                    results = self.retriever.search(self.research_goal, top_k=1)
                    if results:
                        # Try to find a link in the top result
                        links = self.nav.extract_links(results[0]['body'])
                        for l in links:
                            clean_l = l.rstrip(')')
                            if clean_l not in self.visited_urls:
                                target_url = clean_l
                                break

            if not target_url or target_url in self.visited_urls:
                return "No new target URL found or already visited."

            # 2. Scrape
            print(f"Iteration {len(self.iteration_history)+1}: Scraping {target_url}")
            data = self.nav.scrape_url(target_url)
            self.visited_urls.add(target_url)

            if 'error' in data:
                return f"Scrape failed: {data['error']}"

            # 3. Analyze & Discover (Find more links for future iterations)
            new_links = self.nav.extract_links(data['content'])
            for link in new_links:
                clean_link = link.rstrip(')')
                if clean_link not in self.visited_urls and clean_link not in self.link_queue:
                    self.link_queue.append(clean_link)

            # 4. Save/Simulate Analysis
            # In a real scenario, we would process 'data' with an LLM here.
            # For now, we log the success.
            status = f"Successfully processed {data['title']} ({len(data['content'])} chars)"
            self.iteration_history.append({"url": target_url, "status": "success", "title": data['title']})
            return status

        except Exception as e:
            self.iteration_history.append({"url": target_url, "status": "error", "detail": str(e)})
            return f"Iteration Error: {e}"

    def execute_research(self, cycles=3):
        """Main loop to run for a specified number of cycles."""
        print(f"Starting autonomous research for {cycles} cycles...")
        for i in range(cycles):
            result = self.run_iteration()
            print(f"Cycle {i+1} Result: {result}")
            time.sleep(1) # Polite delay
        print("Research execution finished.")

# Initialize and test
controller = AutonomousLoopController(nav, retriever, KB_ROOT)
controller.set_goal("Economic impact of immigration research")
print("AutonomousLoopController defined and initialized.")
```

## Implement Intelligent Link Discovery

### Subtask:
Enhance the `WebNavigator` or `AutonomousLoopController` to prioritize links based on relevance to a 'Research Goal' and handle depth-based crawling.


**Reasoning**:
I will enhance the AutonomousLoopController to include relevance-based link prioritization using the existing TfidfVectorizer and implement depth-based tracking to focus the research crawl.



```python
import numpy as np

class IntelligentLoopController(AutonomousLoopController):
    def __init__(self, nav, retriever, kb_root, max_depth=2):
        super().__init__(nav, retriever, kb_root)
        self.max_depth = max_depth
        # queue stores tuples of (url, depth, score)
        self.link_queue = []

    def calculate_relevance(self, text):
        """Calculates cosine similarity between text snippet and research goal."""
        if not self.research_goal or not self.retriever.vectorizer:
            return 0.0
        try:
            vectors = self.retriever.vectorizer.transform([self.research_goal, text])
            sim = cosine_similarity(vectors[0:1], vectors[1:2])
            return float(sim[0][0])
        except:
            return 0.0

    def add_to_queue(self, url, depth, context_text=""):
        """Adds URL to queue if within depth and not visited, with relevance scoring."""
        clean_url = url.rstrip(')')
        if clean_url not in self.visited_urls and depth <= self.max_depth:
            # Check if already in queue to avoid duplicates
            if not any(item[0] == clean_url for item in self.link_queue):
                score = self.calculate_relevance(context_text)
                self.link_queue.append((clean_url, depth, score))
                # Sort queue by score descending (highest relevance first)
                self.link_queue.sort(key=lambda x: x[2], reverse=True)

    def run_iteration(self):
        """Enhanced iteration with intelligent selection and depth tracking."""
        try:
            target_url = None
            current_depth = 0

            if self.link_queue:
                target_url, current_depth, score = self.link_queue.pop(0)
                print(f"Selected high-relevance link (Score: {score:.3f}): {target_url}")
            else:
                # Fallback to KB search if queue empty
                results = self.retriever.search(self.research_goal, top_k=1)
                if results:
                    links = self.nav.extract_links(results[0]['body'])
                    if links:
                        target_url = links[0].rstrip(')')
                        current_depth = 1

            if not target_url or target_url in self.visited_urls:
                return "No suitable new target URL found."

            print(f"Iteration {len(self.iteration_history)+1} (Depth {current_depth}): Scraping {target_url}")
            data = self.nav.scrape_url(target_url)
            self.visited_urls.add(target_url)

            if 'error' in data:
                return f"Scrape failed: {data['error']}"

            # Discovery: Extract new links and add to intelligent queue
            if current_depth < self.max_depth:
                new_links = self.nav.extract_links(data['content'])
                # Simplified context: taking a snippet around the link would be better,
                # but for now we use the page title/content snippet as context.
                context = f"{data['title']} {data['content'][:200]}"
                for link in new_links:
                    self.add_to_queue(link, current_depth + 1, context)

            status = f"Processed {data['title']} - Discovered {len(new_links)} potential paths."
            self.iteration_history.append({"url": target_url, "depth": current_depth, "status": "success"})
            return status

        except Exception as e:
            return f"Iteration Error: {e}"

# Initialize the intelligent controller
intel_controller = IntelligentLoopController(nav, retriever, KB_ROOT, max_depth=2)
intel_controller.set_goal("Recent developments in nuclear fusion energy")
print("IntelligentLoopController with relevance scoring and depth tracking initialized.")
```
